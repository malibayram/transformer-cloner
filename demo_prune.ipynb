```python
from safetensors.torch import load_file, save_file

d2_model = load_file("embeddinggemmacopy/2_Dense/model.safetensors")
d2_weight = d2_model["linear.weight"]
d2_weight.shape
```

```python
d2_weight = d2_weight[:, :640].contiguous()
print(d2_weight.shape)
new_d2_model = {"linear.weight": d2_weight}
save_file(new_d2_model, "embeddinggemma/2_Dense/model.safetensors")
```

```python
d3_model = load_file("embeddinggemmacopy/3_Dense/model.safetensors")
d3_weight = d3_model["linear.weight"]
d3_weight.shape
```

```python
d3_weight = d3_weight[:640, :].contiguous()
print(d3_weight.shape)
new_d3_model = {"linear.weight": d3_weight}
save_file(new_d3_model, "embeddinggemma/3_Dense/model.safetensors")
```
